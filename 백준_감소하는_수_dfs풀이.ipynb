### dfs를 활용하여 감소하는 수 만들어보기

이전 자리의 숫자에 dependent한 상황이기 때문에, 해당 자리 숫자에서 그 아래 자리 숫자들을 재귀적으로 호출한다.

max_depth를 정해두어, 해당 자리 수의 숫자들이 나올 수 있도록 하고, answer배열을 함께 넣어 호출하여 그동안 쌓였던 숫자들을 호출할 수 있도록 한다. 현재 visited 배열이 같은 dfs, 즉 자릿수가 같고 최대 수가 같은 범위 내(depth 같을 동안)에서는 공유되고 있기 때문에 1)max_depth까지 도달한 경우 2) 자릿수를 채우지 못하고 종료되는 경우에서 다시 reset이 필요하다.

```
global n
n = int(input())
nums = list(range(0,10))
global cnt
cnt = 0

def dfs(v, visited, depth_cnt, depth, answer):
  global cnt
  visited[v]=1
  answer[depth_cnt] = str(v)
  depth_cnt+=1
  if depth_cnt == depth:
    cnt+=1
    print(cnt, ''.join(answer))
    visited[v]=0    # 1)의 경우임. 해당 depth에서 사용한 노드만 visited 복원
    if cnt == n:
      print(''.join(answer))
    return

  else:
    for i in nums:
      if visited[i]==0 and i<v:
        dfs(i, visited, depth_cnt, depth, answer)
    visited[v]=0    # 2)의 경우임. 해당 depth에서 사용한 노드만 visited 복원

for i in range(0,10):
  for j in nums[1:]:
    visited = [0]*len(nums)
    dfs(j,visited,0,i+1,['-1']*(i+1))
if cnt<n:
  print(-1)
```
