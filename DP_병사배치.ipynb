```
def solution(N, arr):
  table = [0]*N
  for i in range(N):
    table[i]=max(arr[i:])
  cnt=0
  for i in range(1,N):
    if arr[i]>=table[i] and arr[i]>arr[i-1]:
      print(arr[i])
      cnt+=1
  return cnt
```

## 병사 배치하기: 문제 해결 아이디어

가장 긴 증가하는 부분 수열: LIS의 변형.
즉, 가장 긴 감소하는 부분 수열을 찾는 문제로 치환 가능

LIS 점화식
 모든 0<=j<i에 대하여, D[i] = max(D[i],D[j]+1) if array[j]<array[i]

 즉, 인덱스 i 위치에 있는 원소를 마지막으로 하면서 increasing 하는 subset의 최대 길이를 table에 업데이트시켜줌

```
def na_solution(N,array):
  array.reverse()  # 순서를 뒤집어 LIS 문제로 치환
  dp = [1]*NameError
  for i in range(1,N):
    for j in range(0,i):
      if array[j]<array[i]:
        dp[i] = max(dp[i],dp[j]+1)
  return N-max(dp)
```
