### 특정 조건을 만족시키는 컬럼 별 개수 구하기

image.png

````
-- 코드를 입력하세요
SELECT animal_type, COUNT(animal_id) AS count FROM animal_ins
WHERE animal_type IN('Cat','Dog')
GROUP BY animal_type
ORDER BY animal_type ASC ;
````

### HAVING

WHERE은 집계함수와 같이 사용할 수 없기 때문에(COUNT 등) GROUP BY 에서 사용하는 조건 구문이라고 생각하면 좋다!

````
-- 코드를 입력하세요
SELECT NAME, COUNT(NAME) 
FROM ANIMAL_INS
GROUP BY NAME
HAVING COUNT(NAME)>1
ORDER BY NAME ASC
````

### 시간대 데이터 집계

```
HOUR( ): DATETIME타입 데이터의 시간 RETURN
```



````
-- 코드를 입력하세요
SELECT HOUR(DATETIME), COUNT(HOUR(DATETIME))
FROM ANIMAL_OUTS
WHERE HOUR(DATETIME) BETWEEN 9 AND 20
GROUP BY HOUR(DATETIME)
ORDER BY HOUR(DATETIME) ASC
````

### 두 테이블을 비교하여 일부 테이블에만 존재하는 데이터 

image.png

image.png

빨간색 선택된 집합의 데이터는 무조건 포함하여 두 테이블을 JOIN할 때, 다른 테이블에서 없는 내용은 NULL로 들어감

````
-- 코드를 입력하세요
SELECT A.ANIMAL_ID, A.NAME
FROM ANIMAL_OUTS A LEFT OUTER JOIN ANIMAL_INS B 
ON A.ANIMAL_ID = B.ANIMAL_ID
WHERE B.ANIMAL_ID IS NULL
ORDER BY A.ANIMAL_ID ASC, A.NAME ASC
````
